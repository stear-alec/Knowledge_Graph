# Import packages

```python
import xml.etree.ElementTree as ET
import pandas as pd
```

# Extract DrugBank XML

```python
file = ET.parse('full_database.xml')
```

```python
root = file.getroot()
```

Finding attributes in DrugBank

```python
for x in root[0]:
    print(x.tag,x.attrib)
```

Finding DrugBank drug name

```python
drugname_from_drugbank = []
for x in root.iter('{http://www.drugbank.ca}name'):
    item = x.text
    drugname_from_drugbank.append(item)
```

Finding DrugBank drug ID

```python
drugbank_id = []
for x in root.findall('{http://www.drugbank.ca}drug'):
    item = x.find('{http://www.drugbank.ca}drugbank-id').text
    drugbank_id.append(item)
```

Finding DrugBank target name

```python
targetname_from_drugbank = []
for x in root.findall('{http://www.drugbank.ca}drug'):
    targets=x.find('{http://www.drugbank.ca}targets')
    for target in targets.findall('{http://www.drugbank.ca}target'):
        item = target.find('{http://www.drugbank.ca}name').text
        targetname_from_drugbank.append(item)
```

Finding DrugBank target ID

```python
targetid_from_drugbank = []
for x in root.findall('{http://www.drugbank.ca}drug'):
    targets=x.find('{http://www.drugbank.ca}targets')
    for target in targets.findall('{http://www.drugbank.ca}target'):
        item = target.find('{http://www.drugbank.ca}id').text
        targetid_from_drugbank.append(item)
```

Associating DrugBank drug and target

```python
l=[]


for Drug in root.findall('{http://www.drugbank.ca}drug'):
    
    DrugID = Drug.find('{http://www.drugbank.ca}drugbank-id').text
    targets = Drug.find('{http://www.drugbank.ca}targets')
    
    for target in targets.findall('{http://www.drugbank.ca}target'):
        target = (target.find('{http://www.drugbank.ca}id').text)
        l.append(["DrugID:", DrugID, "target:", target])
```

Not needed: Finding PharmGKB drug code in DrugBank

```python
drugbank_id = []
pharmgkb_from_drugbank = []

for x in root.findall('{http://www.drugbank.ca}drug'):
    drugbankid = x.find('{http://www.drugbank.ca}drugbank-id').text
    identifiers =x.find('{http://www.drugbank.ca}external-identifiers')
    for  identifier in identifiers.findall('{http://www.drugbank.ca}external-identifier'):
        resource = identifier.find('{http://www.drugbank.ca}resource').text
        if resource == 'PharmGKB':
            item = identifier.find('{http://www.drugbank.ca}identifier').text
            pharmgkb_from_drugbank.append(item)
            drugbank_id.append(drugbankid)
```

Not needed: Finding Uniprot targets(protein) in DrugBank

```python
uniprot_from_drugbank = []
for x in root.findall('{http://www.drugbank.ca}drug'):
    targets = x.find('{http://www.drugbank.ca}targets')
    for target in targets.findall('{http://www.drugbank.ca}target'):
        polypeptide = target.find('{http://www.drugbank.ca}polypeptide')
        if polypeptide != None:
            externalidentifiers= polypeptide.find('{http://www.drugbank.ca}external-identifiers')
            for externalidentifier in externalidentifiers.findall('{http://www.drugbank.ca}external-identifier'):
                if externalidentifier.find('{http://www.drugbank.ca}resource').text == "UniProtKB":
                    item = externalidentifier.find('{http://www.drugbank.ca}identifier').text
                    uniprot_from_drugbank.append(item)
```
